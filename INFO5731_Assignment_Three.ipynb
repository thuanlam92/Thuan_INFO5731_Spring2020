```
#install beautiful soup using pip install bs4 = HTML parsing
#requests = requesting to url and acceessing information from it
import bs4
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup 
import pysbd
from pysbd.utils import PySBDFactory
from nltk.util import ngrams
import collections
from collections import Counter
import nltk
nltk.download('punkt')


my_url = 'https://www.imdb.com/title/tt7286456/reviews?sort=helpfulnessScore&dir=desc&ratingFilter=10'
# my_url = 'https://www.imdb.com/title/tt7286456/reviews?ref_=tt_ql_3'


#opening up connection, grabbing the page and read it
uRev = uReq(my_url) 
page_html = uRev.read()
uRev.close()

#html.parser: to parse as HTML file
page_soup = soup(page_html, "html.parser") 

#grab reviews
reviews = page_soup.findAll("div",{"class":"text show-more__control"})
# reviews

# #get the top 100 reviews
# # https://www.pluralsight.com/guides/web-scraping-with-selenium
# #https://www.accordbox.com/blog/how-crawl-infinite-scrolling-pages-using-python/
# #https://stackoverflow.com/questions/47788605/how-to-use-python-to-press-the-load-more-in-imdb-to-get-more-reviews


# get text
import re
result = re.sub("<.*?>", "", str(reviews)) 
# print(result)

#lowercase texts
lower = result.lower()
# print(lower)


#remove punctuations
import string
translator = str.maketrans('', '', string.punctuation)
lower = lower.translate(translator)
# print(lower)

#remove special characters
char_removed = re.sub(r"[-()\"#/@;:<>{}`+=~|.!?,]", "", str(lower))
#print (char_removed)


#remove numbers
num_removed = re.sub(r'[0-9\.]+', '', str(char_removed))
print (num_removed)
for line in num_removed:
  print(line)


# with open('csvfile.csv','w') as file:
#     # for line in num_removed:
#         writer = csv.writer(file)
#         writer.writerow(('ColA', 'ColB'))
#         file.write(num_removed)
#         # file.write('\n')

# import csv
# number = 1
# outfile = open("./inmates.csv", "w")
# writer = csv.writer(outfile)
# writer.writerow(["Number", "Review", "Sentiment"])
# writer.writerow(["a", "b", "c"])
# # for line in num_removed:
# #   writer.writerows(line)



#tokenization
from nltk.tokenize import RegexpTokenizer 
  
tokenizer = RegexpTokenizer("[\w']+") 
text = tokenizer.tokenize(str(num_removed)) 
# print(text)


# count frequency of all n-grams (n=3)
trigrams = nltk.trigrams(text)
trigrams_freq = nltk.FreqDist(trigrams)
for token in trigrams_freq.keys():
  print(token, trigrams_freq[token])

```

```
#count probability of all bigrams

import pandas as pd
from textblob import TextBlob

f = TextBlob(num_removed)
bigrams = TextBlob(num_removed).ngrams(2)

# bigrams and frequencies
bi_freq = {}
for gram in bigrams:
  gram = tuple(gram)
  bi_freq[gram] = bi_freq.get(gram,0)+1

bi_prob = {}
for k, v in bi_freq.items():
  bi_prob[k] = v/f.words.count(k[0])

df = pd.DataFrame.from_dict(bi_prob,'index', columns = ['bigram probability'])
df
```

```
#collect nouns
import nltk
nltk.download('brown')
from textblob import TextBlob
blob = TextBlob(num_removed)
print(blob.noun_phrases)

no_freq = {}
for np in blob.noun_phrases:
  no_freq[np] = no_freq.get(np,0)+1
no_phrases = [val for val in no_freq.values()] 
max_freq = max([val for val in no_freq.values()])


h = []
for line in num_removed.splitlines():
  blob = TextBlob(line)
  no_phrases = blob.noun_phrases
  h1 = {}
  for np in no_phrases:
    h1[np] = h1.get(np,0)+1
  for k,v in h1.items():
    h1[k] = v/max_freq
  h.append(h1)


index = [n for n in range(1,len(h)+1)]

df = pd.DataFrame(h1, index)
df.fillna(0, inplace=True)
df
```

(40 points). Starting from the documents (all the reviews, or abstracts, or tweets) collected for assignment two, write a python program:

(2) To rank the documents with respect to query (design a query by yourself, for example, "An Outstanding movie with a haunting performance and best character development") by using cosine similarity.

```
#build the documents-terms weights (tf*idf)


from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
import numpy as np
import pandas as pd
import re

# vectorizer = TfidfVectorizer()
# response = vectorizer.fit_transform(text)
# print(response)

# for i, feature in enumerate(vectorizer.get_feature_names()):
#     print(i, feature)

cvec = CountVectorizer(stop_words='english', min_df=3, max_df=0.5, ngram_range=(1,2))
sf = cvec.fit_transform(text)

transformer = TfidfTransformer()
transformed_weights = transformer.fit_transform(sf)
weights = np.asarray(transformed_weights.mean(axis=0)).ravel().tolist()
weights_df = pd.DataFrame({'term': cvec.get_feature_names(), 'weight': weights})

weights_df.sort_values(by='weight', ascending=False).head(10)

#https://iyzico.engineering/how-to-calculate-tf-idf-term-frequency-inverse-document-frequency-from-the-beatles-biography-in-c4c3cd968296
```
