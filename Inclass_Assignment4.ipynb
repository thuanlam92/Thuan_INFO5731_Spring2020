```
import pysbd
import spacy
from pysbd.utils import PySBDFactory

f = open('AdamsTanner.txt','r', encoding='utf8', errors = "ignore")
lines = f.readlines()


#number of sentences
all_sent = []
for line in lines:
  
    # print(line)
    nlp = spacy.blank('en')
    nlp.add_pipe(PySBDFactory(nlp))
    doc = nlp(line)
    # print(doc)
    # print(list(doc.sents))
    for item in list(doc.sents):
      all_sent.append(item)
# print(all_sent)
print(len(all_sent))




#word_list
def convert(lst): 
  all_words = []
  for item in lst:
    if len(item)>1:
      # print(item)
      # print(type(item))
      words = str(item).split()
      for word in words:
        all_words.append(word)
  return all_words

all_word_list = convert(all_sent)
print(all_word_list)    









# #number of characters
len_char = 0

for line in lines:
    count = len(line)
    len_char = len_char+count
print(len_char)



```

```
#average word length

def avg_word(sentence):
  words = str(sentence).split()
  return (sum(len(word) for word in words)/len(words))

average_word = avg_word(all_word_list)
print(average_word)
```

```
#number of uppercase words
#number of special characters
#number of numeric

def Count(str): 
    upper, lower, number, special = 0, 0, 0, 0
    for i in range(len(str)): 
        if str[i] >= 'A' and str[i] <= 'Z': 
            upper += 1
        elif str[i] >= 'a' and str[i] <= 'z': 
            lower += 1
        elif str[i] >= '0' and str[i] <= '9': 
            number += 1
        else: 
            special += 1
    print('Upper case letters:', upper) 
    print('Lower case letters:', lower) 
    print('Number:', number) 
    print('Special characters:', special) 


#conver sentences into a string to count
document = ''

for item in all_word_list:
  document = document + item + " "
print(document)

Count(document)
  
```

```
#lower casing

lower_casing = document.lower()
print(lower_casing)
```

```
#punctuation removal

import string
words = document.split()
table = str.maketrans('', '', string.punctuation)
punctuation_stripped = [w.translate(table) for w in words]
print(punctuation_stripped)
```

```
#number of stopwords
#stopwords removal


import nltk
nltk.download('stopwords')
  
stop_words = set(stopwords.words('english'))
stopword_stripped = [w for w in words if not w in stop_words]
print(stopword_stripped)
```

```

import re
from collections import Counter
#find the 10 most common words
words = re.findall(r'\w+', str(punctuation_stripped))
most_commons = Counter(words).most_common(10)
print(most_commons)

#common words (10 most frequent words) removal
resultwords_1  = [word for word in re.split("\W+",str(punctuation_stripped)) if word.lower() not in most_commons]
print(resultwords_1)





#find the 10 least common words
n=10
least_commons = Counter(words).most_common()[:-n-1:-1] 
print(least_commons)

#rare words (10 least frequent words) removal
resultwords_2  = [word for word in re.split("\W+",str(punctuation_stripped)) if word.lower() not in least_commons]
print(resultwords_2)
```

```
#spelling correction

from textblob import TextBlob

spelling_correction = TextBlob(document).correct()
print(spelling_correction)
```

```
#tokenization


from nltk.tokenize import RegexpTokenizer 
  
tokenizer = RegexpTokenizer("[\w']+") 
text = str(spelling_correction)
tokenizer.tokenize(text) 
```

```
#stemming

from nltk.stem import PorterStemmer 
from nltk.tokenize import word_tokenize 
   
ps = PorterStemmer() 
   
word_stemmed = str(spelling_correction)
for w in words: 
    print(w, " : ", ps.stem(w)) 
```

```
#lemmatization
import nltk
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer 

lemmatizer = WordNetLemmatizer() 

word_lemmatized = str(spelling_correction)
for w in words:
  print(w," : ", lemmatizer.lemmatize(w))

```
