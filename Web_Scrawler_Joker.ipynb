```
#install beautiful soup using pip install bs4 = HTML parsing
#requests = requesting to url and acceessing information from it
import bs4
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup 
import pysbd
import spacy
from pysbd.utils import PySBDFactory

my_url = 'https://www.imdb.com/title/tt7286456/reviews?sort=helpfulnessScore&dir=desc&ratingFilter=10'

#opening up connection, grabbing the page and read it
uRev = uReq(my_url) 
page_html = uRev.read()
uRev.close()

#html.parser: to parse as HTML file
page_soup = soup(page_html, "html.parser") 

#grab reviews
reviews = page_soup.findAll("div",{"class":"text show-more__control"})
reviews

# get text
import re
result = re.sub("<.*?>", "", str(reviews)) 
print(result)
print(type(result))

```

(30 points). Write a python program to clean the text data you collected above and save the data in a new column in the csv file. The data cleaning steps include:
(1) Remove noise, such as special characters and punctuations.
(2) Remove numbers.
(3) Remove stopwords by using the stopwords list.
(4) Lowercase all texts
(5) Stemming.
(6) Lemmatization.

```
#lowercase texts
lower = result.lower()
# print(lower)


#remove punctuations
import string
translator = str.maketrans('', '', string.punctuation)
lower = lower.translate(translator)
# print(lower)

#remove special characters
char_removed = re.sub(r"[-()\"#/@;:<>{}`+=~|.!?,]", "", str(lower))
#print (char_removed)


#remove numbers
num_removed = re.sub(r'[0-9\.]+', '', str(char_removed))
#print (num_removed)



#tokenization
from nltk.tokenize import RegexpTokenizer 
  
tokenizer = RegexpTokenizer("[\w']+") 
text = tokenizer.tokenize(str(num_removed)) 
print(text)



  
```

```
#remove stopwords by using the stopword list

import nltk
# nltk.download('stopwords')
from nltk.corpus import stopwords
  
stop_words = set(stopwords.words('english'))
stopword_stripped = [w for w in text if not w in stop_words]
print(stopword_stripped)
# # #print(type(stopword_stripped))

```

```
#stemming
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

ps = PorterStemmer()

for w in stopword_stripped:
     print(w, " : ", ps.stem(w)) 
```

```
#lemmantation
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer 

lemmatizer = WordNetLemmatizer() 

for w in stopword_stripped:
  print(w," : ", lemmatizer.lemmatize(w))
```

```
#A dependency parse connects words according to their relationships. Each vertex in the tree represents a word, child nodes are words that are dependent on the parent, and edges are labeled by the relationship.
import spacy
from nltk import Tree


en_nlp = spacy.load('en')

doc = en_nlp(num_removed)

def to_nltk_tree(node):
    if node.n_lefts + node.n_rights > 0:
        return Tree(node.orth_, [to_nltk_tree(child) for child in node.children])
    else:
        return node.orth_


[to_nltk_tree(sent.root).pretty_print() for sent in doc.sents]
```
